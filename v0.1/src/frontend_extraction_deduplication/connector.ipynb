```python
# # change the paths according to your files
# # uncomment all the files for which you have predictions/annotations

incidents_df = pd.read_csv('data/incidents.csv')
food_annotations_df = pd.read_csv('data/incidents_csv_food_instafoodroberta.csv')
#hazards_annotations_df = pd.read_csv('data/hazards_annotations.csv')
#locations_annotations_df = pd.read_csv('data/locations_annotations.csv')
#organizations_annotations_df = pd.read_csv('data/organizations_annotations.csv')
```

```python
# rename column "Unnamed: 0" to "incident_id"
incidents_df.rename(columns = {'Unnamed: 0':'incident_id'}, inplace = True)
```

```python
# group by text_id and create a list of phrase
food_annotations_df = food_annotations_df.groupby('text_id').agg({'phrase': set}).reset_index()
# convert the set to a list of phrases
food_annotations_df['phrase'] = food_annotations_df['phrase'].apply(lambda x: list(x))
```

```python
# merge the incidents_df with the food_annotations_df on incident_id = text_id
incidents_df = incidents_df.merge(food_annotations_df, left_on='incident_id', right_on='text_id', how='left')
# drop the text_id column
incidents_df.drop(columns=['text_id'], inplace=True)
# rename the phrase column to food
incidents_df.rename(columns = {'phrase':'foods_mistral'}, inplace = True)
```
